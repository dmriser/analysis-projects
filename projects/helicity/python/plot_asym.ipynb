```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def load_asym(filename=None):
    run = []
    value = []
    error = []
    chi2 = []
    
    with open(filename) as f:
        for line in f:
            run.append(line.split(',')[0])
            value.append(line.split(',')[1])
            error.append(line.split(',')[2])
            chi2.append(line.split(',')[3])
    
    unsorted_run = np.array(run, dtype=np.int)
    unsorted_val = np.array(value, dtype=np.float)
    unsorted_err = np.array(error, dtype=np.float)
    unsorted_chi2 = np.array(chi2, dtype=np.float)

    # sort 
    p = unsorted_run.argsort()
    run = unsorted_run[p]
    val = unsorted_val[p]
    err = unsorted_err[p]
    chi2 = unsorted_chi2[p]


    return run, val, err, chi2

def filter_between(data=None, filter_min=-1.0, filter_max=1.0):
    return data[(data > filter_min) & (data < filter_max)]
```

```python
run, all_val, all_err, all_chi2 = load_asym('asym_nowp.dat')
run_wp, all_val_wp, all_err_wp, all_chi2_wp = load_asym('asym.dat')
run_uncorr, all_val_uncorr, all_err_uncorr, all_chi2_uncorr = load_asym('asym_uncorr.dat')

val_uncorr = filter_between(all_val_uncorr, -0.06, 0.06)
val = filter_between(all_val, -0.06, 0.06)
val_wp = filter_between(all_val_wp, -0.06, 0.06)

err_uncorr = filter_between(all_err_uncorr, -0.01, 0.03)
err = filter_between(all_err, -0.01, 0.03)
err_wp = filter_between(all_err_wp, -0.01, 0.03)

print('Mean Asymmetry = %f, Std. = %f') % (np.mean(val), np.std(val))
print('Mean Asymmetry = %f, Std. = %f') % (np.mean(val_wp), np.std(val_wp))
print('Mean Asymmetry (uncorr) = %f, Std. = %f') % (np.mean(val_uncorr), np.std(val_uncorr))
```

```python
sns.set_style('whitegrid')

axis = sns.distplot(val_wp, bins=60, label='Corrected')
sns.distplot(val_uncorr, color='r', bins=60, label='Uncorrected')
axis.set(xlabel='$A_{LU}^{sin\phi}$', ylabel='Counts')
axis.legend()
fig = axis.get_figure()
fig.savefig('asym_histo.png')
```

```python
axis = sns.distplot(all_chi2_wp, bins=60, label='Corrected')
#sns.distplot(all_chi2_uncorr, color='r', bins=60, label='Uncorrected')
axis.set(xlabel='$\chi^2 /NDF$', ylabel='Counts')
axis.legend()
fig = axis.get_figure()
fig.savefig('chi2_histo.png')
```

```python
sns.set_style('whitegrid')

axis = sns.distplot(err_wp, bins=60, label='Corrected')
sns.distplot(err_uncorr, color='r', bins=60, label='Uncorrected')
axis.set(xlabel='$\delta A_{LU}^{sin\phi}$', ylabel='Counts')
axis.legend()
fig = axis.get_figure()
fig.savefig('asym_err_histo.png')
```

```python
axis = sns.distplot(all_chi2_uncorr, color='r', bins=60, label='Uncorrected')
axis.set(xlabel='$\chi^2 /NDF$', ylabel='Counts')
axis.legend()
fig = axis.get_figure()
fig.savefig('chi2_uncorr_histo.png')
```
